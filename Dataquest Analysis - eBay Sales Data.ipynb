```python
brand_types = autos.brand.value_counts(normalize = True)
common_brands = brand_types[brand_types > 0.05].index
print(common_brands)
```

### Description of Brand Data and Aggregation Decision

The data represents the distribution of the brands in the dataset. The top 3 are volkswagen, bmw, and opel.

I've chosen to aggregate the dataset by cars that take up atleast 5% of the dataset, as they offered as a recommendation. 

```python
aggregated_brands = {}


for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    aggregated_brands[brand] = int(mean_price)
    
print(aggregated_brands)
```

### Analysis of the above results

The Audi has the highest mean price, followed by BMW, and then Ford. 


```python
average_mileage = {}
for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    avg_mileage = brand_only["odometer_km"].mean()
    average_mileage[brand] = int(avg_mileage)
    
print(average_mileage)
```

```python
avg_price = pd.Series(aggregated_brands)
print(avg_price)
```

```python
avg_mileage = pd.Series(average_mileage)
print(avg_mileage)
```

```python
df = pd.DataFrame(avg_price, columns = ['mean_price'])
print(df)
```

```python
df['average_mileage'] = avg_mileage
```

```python
df
```

### Analysis of the above

The car with the highest average mileage are bmw, mercedez_benz, and Audi. These cars also have the highest prices in the dataset. It seems then that average_mileage does not negatively effect the mean_price of the car. You could technically make the argument, based on this data, that a higher average_mileage will therefore make the price of the car increase in value. However, we know that not to be the case in the real world.

I would then draw the assumption that the mean_price is more affected by the brand of the car rather than the mileage.
